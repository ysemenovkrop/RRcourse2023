```python
# Importing the libraries
import os
import pandas as pd

# Specifying the directory
os.chdir("/Users/yuriisemenov/Documents/GitHub/RRcourse2023/6. Coding and documentation/Data")

# Importing all sheets in once
excel_file = pd.read_excel("Eurostat_employment_isco.xlsx", sheet_name=None)

isco1 = excel_file["ISCO1"]
isco2 = excel_file["ISCO2"]
isco3 = excel_file["ISCO3"]
isco4 = excel_file["ISCO4"]
isco5 = excel_file["ISCO5"]
isco6 = excel_file["ISCO6"]
isco7 = excel_file["ISCO7"]
isco8 = excel_file["ISCO8"]
isco9 = excel_file["ISCO9"]

# Calculating sums
total_Belgium = isco1["Belgium"] + isco2["Belgium"] + isco3["Belgium"] + isco4["Belgium"] + isco5["Belgium"] + isco6["Belgium"] + isco7["Belgium"] + isco8["Belgium"] + isco9["Belgium"]
total_Spain = isco1["Spain"] + isco2["Spain"] + isco3["Spain"] + isco4["Spain"] + isco5["Spain"] + isco6["Spain"] + isco7["Spain"] + isco8["Spain"] + isco9["Spain"]
total_Poland = isco1["Poland"] + isco2["Poland"] + isco3["Poland"] + isco4["Poland"] + isco5["Poland"] + isco6["Poland"] + isco7["Poland"] + isco8["Poland"] + isco9["Poland"]
```

```python
import pandas as pd

dfs = [isco1, isco2, isco3, isco4, isco5, isco6, isco7, isco8, isco9]
for i, df in enumerate(dfs):
    df['ISCO'] = i + 1
all_data = pd.concat(dfs, ignore_index=True)

print(type(all_data))

totals = {'Belgium': total_Belgium, 'Spain': total_Spain, 'Poland': total_Poland}
for country, total in totals.items():
    all_data[f"total_{country}"] = total * 9

all_data['share_Belgium'] = all_data['Belgium'] / all_data['total_Belgium']
all_data['share_Spain'] = all_data['Spain'] / all_data['total_Spain']
all_data['share_Poland'] = all_data['Poland'] / all_data['total_Poland'] 

print(all_data)
```
